### Task 2
Test the hypothesis that the offensive productions stats of the Cleveland Cavaliers and Golden State Warriors (the teams that met in the finals that year) were from the same distribution in the 2015/2016 season.

Offensive production refers to two variables: **PTS (Points)** and **FG_PCT (Field Goal Percentage)**. We will need to do two separate hypothesis tests, one for each variable.

```python
team_list = ['Cleveland Cavaliers','Golden State Warriors']

filtered_season15=season15[season15['TEAM_NAME'].isin(team_list)]

cleveland = season15[season15['TEAM_NAME'].str.contains('Cleveland Cavaliers')]
golden = season15[season15['TEAM_NAME'].str.contains('olden State Warriors')]
print(cleveland)
print(golden)
```

```python
t_value,p_value=st.ttest_ind(cleveland["PTS"],golden["PTS"])
significant_level=0.05
print('Test statistic is %f'%float("{:.6f}".format(t_value)))
print('alpha =',significant_level)
print('p-value for two tailed test is %f'%p_value)


if p_value<=significant_level:

    print('reject null hypothesis')
else:

    print('accept null hypothesis')
```

### Task 3
Test the hypothesis that the number of points (PTS) scored by Cleveland Cavaliers changed significantly after the head coach changed in the 2015/2016 season.

- **Coach Blatt was fired on 24th of Jan, 2016**. 

Use the data from seasons 2014/2015 and 2015/2016 - those are years when Cleveland was coached by Blatt. 

**We have two possible solutions to try here:**
- Take the same amount of games from before and after and try t-test.
- Take all the games from before and after and look for the right test to compare two samples with different sizes. (You will need to go through the scipy documentation or google to figure out what kind of test is required.)

```python
#Coach Blatt was fired on 24th of Jan, 2016. take data from season 14-15 and 15-16( cut at jan 24 2016) and compare it with AFTER
```

```python
cleveland13 = season13[season13['TEAM_NAME'].str.contains('Cleveland Cavaliers')]
cleveland14 = season14[season14['TEAM_NAME'].str.contains('Cleveland Cavaliers')]
cleveland15 = season15[season15['TEAM_NAME'].str.contains('Cleveland Cavaliers')]

frames = [cleveland14, cleveland15]

result = pd.concat(frames)

# start_date = "2019-1-1"
cut_date= "2016-01-24"

with_coach_blatt = result[result["GAME_DATE"] < cut_date].sort_values(['GAME_DATE'], ascending=False).head(40)

without_coach_blatt= result[result["GAME_DATE"] > cut_date]

# print(with_coach_blatt)
# print(without_coach_blatt)
print(with_coach_blatt.shape)
print(without_coach_blatt.shape)
```

```python
t_value,p_value=st.ttest_ind(with_coach_blatt["PTS"],without_coach_blatt["PTS"])
significant_level=0.05
print('Test statistic is %f'%float("{:.6f}".format(t_value)))
print('alpha =',significant_level)
print('p-value for two tailed test is %f'%p_value)


if p_value<=significant_level:

    print('reject null hypothesis')
else:

    print('accept null hypothesis')
```
